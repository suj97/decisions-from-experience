### Sensitivity analysis

#### Diff. no. layers

```python
outfile = open('results/MLP/diff-no-layers.txt', 'w')
no_layers = [1,2,3,5,7]

d = 0.2

for layers in no_layers:
    model = Sequential()
    if(layers == 1):
        model.add(Dense(100, activation='relu', kernel_initializer='normal', input_dim = input_length))
        model.add(Dropout(d))
    else:
        while(layers):
            model.add(Dense(100, activation='relu', kernel_initializer='normal'))
            model.add(Dropout(d))
            layers -= 1
    model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))

    opt = optimizers.Adam()
    model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100, batch_size=10)
    
    scores = model.evaluate(X_test, y_test, verbose=0)
    print("Accuracy: %.2f%%" % (scores[1]*100))
    scores1 = model.evaluate(X_train, y_train, verbose=0)
    print("Accuracy: %.2f%%" % (scores1[1]*100))
    outfile.write(str((layers, scores[1]*100, scores1[1]*100)))
    outfile.write('\n')
outfile.close()
```

#### Diff. no units

```python
outfile = open('results/MLP/diff-no-units.txt', 'w')
no_units = [5,10,50,100,500]

for units in no_units:
    model = Sequential()
    model.add(Dense(units, activation='relu', kernel_initializer='normal', input_dim = input_length))
    model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))

    opt = optimizers.Adam()
    model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100, batch_size=20)
    
    scores = model.evaluate(X_test, y_test, verbose=0)
    print("Accuracy: %.2f%%" % (scores[1]*100))
    scores1 = model.evaluate(X_train, y_train, verbose=0)
    print("Accuracy: %.2f%%" % (scores1[1]*100))
    outfile.write(str((units, scores[1]*100, scores1[1]*100)))
    outfile.write('\n')
outfile.close()
```

#### Diff. no epochs

```python
outfile = open('results/MLP/diff-no-epochs.txt', 'w')
no_epochs = [10,50,100,200,500]

for epochs in no_epochs:
    model = Sequential()
    model.add(Dense(100, activation='relu', kernel_initializer='normal', input_dim = input_length))
    model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))

    opt = optimizers.Adam()
    model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=epochs, batch_size=20)
    
    scores = model.evaluate(X_test, y_test, verbose=0)
    print("Accuracy: %.2f%%" % (scores[1]*100))
    scores1 = model.evaluate(X_train, y_train, verbose=0)
    print("Accuracy: %.2f%%" % (scores1[1]*100))
    outfile.write(str((epochs, scores[1]*100, scores1[1]*100)))
    outfile.write('\n')
outfile.close()
```

```python
outfile = open('results/MLP/diff-no-bs.txt', 'w')
no_bs = [1,5,10,50,80]

for bs in no_bs:
    model = Sequential()
    model.add(Dense(100, activation='relu', kernel_initializer='normal', input_dim = input_length))
    model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))

    opt = optimizers.Adam()
    model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100, batch_size=bs)
    
    scores = model.evaluate(X_test, y_test, verbose=0)
    print("Accuracy: %.2f%%" % (scores[1]*100))
    scores1 = model.evaluate(X_train, y_train, verbose=0)
    print("Accuracy: %.2f%%" % (scores1[1]*100))
    outfile.write(str((bs, scores[1]*100, scores1[1]*100)))
    outfile.write('\n')
outfile.close()
```
